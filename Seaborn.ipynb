# Line plot


```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
var = [1,2,3,4,5,6,7]
var_1 = [2,3,4,1,5,6,7]
```

```python
plt.plot(var , var_1)
plt.show()
```

```python
import pandas as pd
```

```python
x_1 = pd.DataFrame({"var" : var , "var_1" : var_1})
```

```python
sns.lineplot(x = "var" , y = "var_1" , data = x_1)
plt.show()
```

```python
sns.lineplot(x = var , y = var_1)
plt.show()
```

```python
y_1 = sns.load_dataset("penguins")
```

```python
y_1
```

```python
sns.lineplot(x = "bill_length_mm" , y = "bill_depth_mm" , c = 'g', data = y_1)
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm" , y = "bill_depth_mm" , c = 'g', data = y_1 , hue = 'sex')
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm" , y = "bill_depth_mm" , size = 10 , c = 'g', data = y_1 , hue = 'sex')
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm" , y = "bill_depth_mm" , style = 'sex' , c = 'g', data = y_1 , hue = 'sex')
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm" , y = "bill_depth_mm" , palette = "Accent" , style = 'sex' , c = 'g', data = y_1 , hue = 'sex')
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm" , y = "bill_depth_mm" , size = 80 , markers = ["o" , "<"], palette = "Accent" , style = 'sex' , c = 'g', data = y_1 , hue = 'sex')
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm" , y = "bill_depth_mm" , dashes = False , size = 80 , markers = ["o" , "<"], palette = "Accent" , style = 'sex' , c = 'g', data = y_1 , hue = 'sex')
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm" , y = "bill_depth_mm" , legend = False , dashes = False , size = 80 , markers = ["o" , "<"], palette = "Accent" , style = 'sex' , c = 'g', data = y_1 , hue = 'sex')
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm" , y = "bill_depth_mm" , legend = False , dashes = False , size = 80 , markers = ["o" , "<"], palette = "Accent" , style = 'sex' , c = 'g', data = y_1 , hue = 'sex')
plt.grid()
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm" , y = "bill_depth_mm" , legend = False , dashes = False , size = 80 , markers = ["o" , "<"], palette = "Accent" , style = 'sex' , c = 'g', data = y_1 , hue = 'sex')
plt.grid()
plt.title("python")
plt.show()
```
