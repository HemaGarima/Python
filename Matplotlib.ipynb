## Simple Bar chart showing revenues of major US tech companies

```python
company = ['GOOGL' , 'AMZN' , 'MSFT' , 'FB']
revenue = [90 , 136 , 89 , 27]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos , revenue , label = "Revenue")
plt.xticks(xpos , company)
plt.ylabel("Revenue(Bln)")
plt.xlabel("Companies")
plt.title('US Technology Stocks')
plt.legend()
```

### Multiple bars showing revenue and profit of major US tech companies

```python
profit = [40 , 2 , 34 , 12]
```

```python
plt.bar(xpos-0.2 , revenue , width = 0.4 , label = "Revenue")
plt.bar(xpos+0.2 , profit , width = 0.4 , label = "Profit")
plt.xticks(xpos , company)
plt.ylabel("Revenue(Bln)")
plt.xlabel("Companies")
plt.title("US Technology Stocks")
plt.legend()
```

### Horizontal Bar chart using barh function

```python
plt.barh(xpos , revenue , label = "Revenue")
plt.yticks(xpos , company)
plt.ylabel("Companies")
plt.xlabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

# Histogram

### In histograms , x axis contains a variable and y axis will be a frequency of that variable

- We have a simple data of blood sugar level of different patients , we will try to plot number of patients by blood range and try to figure out how many patients are normal , pre-diabetic and diabetic

```python
blood_sugar = [113 , 85 , 90 , 150 , 149 , 88 , 93 , 115 , 135 , 80 , 77 , 82 , 129]
```

```python
plt.hist(blood_sugar , rwidth = 0.8) # by default number of bins is set to 10
```

```python
plt.hist(blood_sugar , rwidth = 0.5 , bins = 4) # Divides the data into 4 bins (intervals), meaning the range of values in blood_sugar will be split into 4 groups.
```

### Histogram showing normal , pre-diabetic and diabetic patients distribution



*   80 - 100 : Normal
*   100 - 125 : Pre-diabetic
*   125 - 150 : Diabetic



```python
plt.xlabel("Sugar Level")
plt.ylabel("Number of Patients")
plt.title("Blood Sugar Chart")
plt.hist(blood_sugar , bins = [80 , 100 , 125 , 150] , rwidth = 0.95 , color = 'g')
```

### Multiple data samples in a histogram

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number of Patients")
plt.title("Blood Sugar Chart")
blood_sugar_men = [113 , 85 , 90 , 150 , 149 , 88 , 93 , 115 , 135 , 80 , 77 , 82 , 129]
blood_sugar_women = [67 , 98 , 89 , 120 , 133 , 150 , 84 , 69 , 89 , 79 , 120 , 112 , 100]
plt.hist([blood_sugar_men , blood_sugar_women] , bins = [80 , 100 , 125 , 150] , rwidth = 0.8 , color = ['green' , 'orange'] , label = ['men' , 'women'])
plt.legend()
```

### histtype = step

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number of Patients")
plt.title("Blood Sugar Chart")
plt.hist(blood_sugar , bins = [80 , 100 , 125 , 150] , rwidth = 0.95 , color = 'g' , histtype = 'step')
```

### horizontal orientation

```python
plt.ylabel("Sugar Level")
plt.xlabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,orientation = 'horizontal')
```

# Pie - Charts

```python
exp_values = [1400 , 600 , 300 , 410 , 250]
exp_labels = ["Home Rent" , "Food" , "Phone/Internet Bill" , "Car" , "Other Utilities"]
plt.pie(exp_values , labels = exp_labels)
```

## Draw pie chart as perfect circle

```python
plt.pie(exp_values , labels = exp_labels , shadow = True)
plt.axis("equal")
plt.show()
```

### Show percentages for every pie. Specify radius to increase chart size

```python
plt.axis("equal")
plt.pie(exp_values , labels = exp_labels , shadow = True , autopct = '%1.1f%%' , radius = 1.5)
plt.show()
```

### Explode

```python
plt.axis("equal")
plt.pie(exp_values , labels = exp_labels , shadow = True , autopct ='%1.1f%%' , radius = 1.5 , explode=[0,0,0,0.1,0.2])
plt.show()
```

## Counterclock and angle properties

```python
plt.axis("equal")
plt.pie(exp_values , labels = exp_labels , shadow = True , autopct = '%1.1f%%' , explode=[0,0,0,0.1,0.2],counterclock=False, startangle=45)
plt.show()
```

## Saving chart to a file using savefig

```python
plt.axis("equal")
plt.pie(exp_values , labels = exp_labels , radius = 2 , autopct='%0.1f%%', explode=[0,0.1,0.1,0,0])
plt.savefig("piechart.jpg" , bbox_inches = "tight" , pad_inches = 1 , transparent = True)
```
